```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
korearepub=pd.read_excel("P_Popular Indicators.xlsx",sheet_name="Korea Republic") 
```

```python
koreamacros=korearepub.iloc[:,[2,25,28,29,32,33,57,58,54,51]]
```

```python
koreamacros.iloc[:,[2,9,0,3]].mean()
```

```python
#Null- There is no significant difference in averages of all variables.
#Alt - There is significant difference in averages of all variables.
```

```python
from scipy.stats import f_oneway
```

```python
f_oneway(koreamacros.iloc[:,0],koreamacros.iloc[:,2],koreamacros.iloc[:,3],koreamacros.iloc[:,9])
```

```python
#since pvalue=2.94662067801933e-09 is less than 0.05 ,reject null hypothesis
```

```python
#	Univariate Time Series Forecasting for 2 variables - Inflation, consumer prices (annual %) & - GDP growth (annual %)
#Augmented Dickey Fuller Test of Stationarity
#Null-Unit root present or data is not stationary
#Alt-No unit root present or data is stationary
```

```python
from statsmodels.tsa.stattools import adfuller 
```

```python
adfuller(koreamacros.iloc[:,9])
#since 0.4986502260035451 is less than 0.05,reject null hypothesis
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
seasonal_decompose(koreamacros.iloc[:,9],period=12).plot()
```
